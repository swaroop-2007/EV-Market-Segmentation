```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("EV Stats-1.csv")
df=df.drop('Sl. No',axis=1)
df.head()
```

GRAPH FOR TOP 10 States Which Has Highest Total Number of EV

```python
newdf=df.sort_values('Total in state',ascending=False)
newdf.reset_index()
newdf=newdf.reset_index(drop=True)
newdf.head()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x='State',y="Total in state",data=newdf.iloc[:10]).set(title='TOP 10 States Which has Highest Total No. of E.V')
```
